Caso quiséssemos criar gráficos para fazer comparação entre os estados da mesma região, seria necessário criar uma nova coluna com a região de cada estado antes disso.

```python
# separar o indicador de reião presente no nome de cada estado
dados_usaveis['Regiao'] = dados_usaveis.index.str[0]

# criar um dicionário com as regiões do pais
regioes = {1:'N', 2:'NE', 3:'SE', 4:'S', 5:'CO'}

# aplicar o código criado no dicionário para cada estado
dados_usaveis['Regiao'] = dados_usaveis['Regiao'].map(lambda i: regioes[int(i)])
dados_usaveis['Regiao']
```

Agora podemos escolher uma região qualquer e plotar os gastos apenas para aquela região:

```python
dados_usaveis[dados_usaveis['Regiao']=='SE'].drop('Regiao', axis=1).T.plot(figsize=(15,5))
```

Nesse gráfico, além de transpor as linhas, como fizemos anteriormente, também tivemos que eliminar (**drop**) a coluna "**Região**", pois esta não era uma coluna numérica (além de não gerar informação útil para o gráfico).

Outra opção de seleção seria plotar apenas os estados que tiveram os maiores gastos no período. Para isso, temos que criar uma coluna com a soma total dos gastos e ordenar por esta.

```python
# criar a coluna de soma dos gastos no período
dados_usaveis['Total'] = dados_usaveis.sum(axis=1)
dados_usaveis['Total']
```

```python
# ordenar os dados de forma decrescente 
dados_usaveis['Total'].sort_values(ascending=False)
```

Com os dados organizados, podemos fazer um gráfico bem bonito!

```python
# selecionar apenas os 5 primeiros estados
maiores_gastos = dados_usaveis['Total'].sort_values(ascending=False)[:5].index

#gerar o plot a partir do Data Frame
ax = dados_usaveis.T[maiores_gastos].drop(['Regiao',
                                           'Total']).plot(figsize=(15,5),
                                                                   legend=False)

# inserir um título 
plt.title('Estados com Maiores Gastos com Procedimentos Hospitalares no SUS', fontsize=20)
# configurar o formato dos números do eixo "y"
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
# ajustar a legenda para conter apenas o nome dos estados
legenda = [val[2:] for val in maiores_gastos]
plt.legend(legenda, fontsize=12, frameon=False, ncol=5, loc='upper center')
```

    ref.: https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o

```python
habitantes=pd.read_csv('população estados.csv', sep='	', index_col=0)
habitantes.head()
```

```python
habitantes.info()
```

Os dados de População presentes nesse novo dataframe não são valores do tipo **int64**. Além disso, há um espaço de separação entre os milhares. Para poder usar esses dados, precisamos primeiro eliminar esses espaços.

```python
habitantes=pd.read_csv('população estados.csv', sep='	', index_col=1,
                      decimal=',', thousands=' ')
habitantes.index = habitantes.index.str.strip()
habitantes.head()
```

```python
habitantes.info()
```

Agora temos a informação desejada no formato que conseguimos trabalhar. Precisamos fazer com que as duas tabelas se "conversem":

```python
# Renomear os estados da federação para não aparecer os números na frente
dados_usaveis_mod = dados_usaveis.copy()
dados_usaveis_mod.index = dados_usaveis_mod.index.str[3:]
dados_usaveis_mod.head()
```

```python
dados_usaveis_mod['População'] = habitantes['População']
```

Agora que temos a coluna de "População" no nosso dataframe, podemos criar uma nova de "Gastos/Habitante" e plotar esta para poder avaliar melhor.

```python
# Calcular os gastos por habitante
dados_usaveis_mod['Gastos/Habitante'] = dados_usaveis_mod['Total']/dados_usaveis_mod['População']

# ordenar os gastos por habitante e plotar 
ax = dados_usaveis_mod['Gastos/Habitante'].sort_values(ascending=False).plot(
                                                                             figsize=(12,8), kind='bar',
                                                                             )
# Ajustar grid
ax.set_yticks(np.arange(0,1201,50), minor=True)
ax.grid(which='minor', axis='y', linestyle='--')
ax.grid(which='major', axis='y', linewidth=1, color='darkgrey')
# ajustar limite o eixo "y"
ax.set_ylim(0,1200)
# Nomear eixos e gráfico
ax.set_title('''Gasto Hospitalares por Habitante 
             ''', fontsize=20)
ax.set_xlabel('Unidade da Federação', fontsize=15)
ax.set_ylabel('Gasto por habitante [R$/Hab.]', fontsize=15)

# ref.: https://stackoverflow.com/questions/1388450/giving-graphs-a-subtitle-in-matplotlib
plt.figtext(0.4, 0.895,'(02/2008 - 06/2020)', fontsize=15)
```

Em uma olhada rápida dos dados é possível perceber que os estados que apresentam maior gasto por habitante são os estados do Sul, que são conhecidos por possuir maior qualidade de vida.

Vamos verificar melhor como está essa divisão entre as regiões:

```python
import seaborn as sns
ordem = dados_usaveis_mod['Gastos/Habitante'].sort_values(ascending=False).index

plt.figure(figsize=(12,8))
sns.barplot(data=dados_usaveis_mod, 
            x=dados_usaveis_mod.index,
            y=dados_usaveis_mod['Gastos/Habitante'], 
            order= ordem,
           hue=dados_usaveis_mod['Regiao'],
           )
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.show()
```

Conforme disconfiamos, os estados do Sul têm os maiores gastos por habitante. Em contra partida os estados do N apresentam os menores gastos.

```python
# Ler tabela de IDH
idh = pd.read_csv('IDH_estados_BR.csv', sep='\t', index_col=1,
                 skiprows=1, decimal=',')
# Eliminar espaços brancos antes e depois do nome de cada estado
idh.index = idh.index.str.strip()

# Incluir os dados de IDH no dataFrame dados_usaveis_mod
idh_2017 = idh['IDH']
dados_usaveis_mod['IDH 2017'] = idh_2017

# Plotar gráfico
plt.figure(figsize=(8,8))
ax =sns.scatterplot(
                    x=dados_usaveis_mod['Gastos/Habitante'],
                    y=dados_usaveis_mod['IDH 2017'],
                    hue=dados_usaveis_mod['Regiao'],
                    #legend = False
                )
# Ajustar grid do plot
# ref. https://stackoverflow.com/questions/52260191/matplotlib-set-minor-ticks-under-seaborn
sns.set(rc={"xtick.bottom" : True, "ytick.left" : True})
#ax.set_xticks(np.arange(0,1201,25), minor=True)
ax.set_yticks(np.arange(0,1.01,.01), minor=True)
ax.grid(which='minor', axis='y')
ax.grid(which='major', axis='y', color='darkgray')
ax.grid(which='minor', axis='x')
ax.grid(which='major', axis='x',color='darkgray')
ax.set_title('Influência dos Gastos Totais em Saude Hospitalar no IDH de Cada Estado \n', fontsize=20)
ax.set_xlabel('Gastos em Saúde Hospitalar', fontsize=15)
ax.set_ylabel('IDH 2017', fontsize=15)
plt.ylim(0.6,.9)
plt.xlim(300,1200)
# Ajustar legenda
plt.legend(loc='upper center', ncol=6)

ax.text(dados_usaveis_mod.loc['Ceará','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Ceará','IDH 2017'],
        'Ceará')
ax.text(dados_usaveis_mod.loc['Rio de Janeiro','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Rio de Janeiro','IDH 2017'],
        'Rio de Janeiro')
ax.text(dados_usaveis_mod.loc['Bahia','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Bahia','IDH 2017'],
        'Bahia')
ax.text(dados_usaveis_mod.loc['São Paulo','Gastos/Habitante'], 
        dados_usaveis_mod.loc['São Paulo','IDH 2017'],
        'São Paulo')
ax.text(dados_usaveis_mod.loc['Amazonas','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Amazonas','IDH 2017'],
        'Amazonas')
ax.text(dados_usaveis_mod.loc['Paraná','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Paraná','IDH 2017'],
        'Paraná')
ax.text(dados_usaveis_mod.loc['Distrito Federal','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Distrito Federal','IDH 2017'],
        'Distrito Federal')
ax.text(dados_usaveis_mod.loc['Amapá','Gastos/Habitante'], 
        dados_usaveis_mod.loc['Amapá','IDH 2017'],
        'Amapá')
plt.show()
```
