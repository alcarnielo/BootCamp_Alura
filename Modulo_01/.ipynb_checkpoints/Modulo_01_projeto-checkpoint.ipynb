#### Escolha das colunas para análise

```python
#### Incluir a coluna 'place_type'
# nos dados por estado
cols_estados.append('place_type')
dados_estados = covid19_2[cols_estados]
# nos dados por municipio
cols_municipios.append('place_type')
dados_municipios = covid19_2[cols_municipios]
```

```python
mortes_por_estado = dados_estados[dados_estados['place_type']=='state'].groupby(
    'state')['new_deaths'].sum().sort_values(ascending=False)
mortes_por_municipio = dados_estados[dados_estados['place_type']=='city'].groupby(
    'state')['new_deaths'].sum().sort_values(ascending=False)
mortes =pd.DataFrame()
#mortes['Ref.'] = referencia.loc['SP':,'Mortes']
#mortes['Por Estado'] = mortes_por_estado
#mortes['Por Município'] = mortes_por_municipio
mortes['Diferença Estado - Municipio'] = (abs(mortes_por_estado - mortes_por_municipio))
                                          
#mortes['Diferença Estado - Ref.'] = mortes_por_estado - referencia.loc['SP':,'Mortes']
#mortes['Diferença Municipio - Ref.'] = mortes_por_municipio - referencia.loc['SP':,'Mortes']
ax = mortes[mortes['Diferença Estado - Municipio']
            > 0].sort_values('Diferença Estado - Municipio').plot.barh(figsize=(10,6), )
ax.set_title('''Mortes no estado - Mortes nos municípios
''', fontsize=20)
ax.set_xlabel('Diferença', fontsize=15)
ax.get_legend().set_visible(False)
plt.text(100,1,'Estadoos com diferença igual a zero não são mostrados')
```

```python
mortes_por_estado = dados_estados[dados_estados['place_type']=='state'].groupby(
    'state')['new_deaths'].sum().sort_values(ascending=False)
mortes_por_municipio = dados_estados[dados_estados['place_type']=='city'].groupby(
    'state')['new_deaths'].sum().sort_values(ascending=False)
mortes =pd.DataFrame()
#mortes['Ref.'] = referencia.loc['SP':,'Mortes']
#mortes['Por Estado'] = mortes_por_estado
#mortes['Por Município'] = mortes_por_municipio
mortes['Diferença Estado - Municipio'] = (abs(mortes_por_estado - mortes_por_municipio)
                                          /referencia.loc['SP':,'Mortes']*100)
#mortes['Diferença Estado - Ref.'] = mortes_por_estado - referencia.loc['SP':,'Mortes']
#mortes['Diferença Municipio - Ref.'] = mortes_por_municipio - referencia.loc['SP':,'Mortes']
ax = mortes[mortes['Diferença Estado - Municipio']
            > 0].sort_values('Diferença Estado - Municipio').plot.barh(figsize=(10,6), )
ax.set_title('''Mortes no estado - Mortes nos municípios
---------------------------------------------------------------
Mortes conforme Google
''', fontsize=20)
ax.set_xlabel('Diferença [%]', fontsize=15)
ax.get_legend().set_visible(False)
plt.text(10,1,'Estadoos com diferença igual a zero não são mostrados')
```
