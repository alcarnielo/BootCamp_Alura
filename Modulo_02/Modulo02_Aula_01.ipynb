```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
fonte = 'https://raw.githubusercontent.com/alcarnielo/Bootcamp_Alura/main/Modulo_02/A200857189_28_143_208.csv'
dados = pd.read_csv(fonte, sep=';', encoding='ISO-8859-1', index_col=0,
                    skiprows=3, skipfooter=12, 
                    thousands='.', decimal=',', na_values='-')
dados.head()
```

```
gastos_mais_recentes = dados['2020/Ago']
gastos_mais_recentes.sort_values(ascending=False).plot.bar(figsize=(12,8))
```

```
(gastos_mais_recentes/gastos_mais_recentes.sum()).sort_values(ascending=False).plot(figsize=(12,8),kind='bar')
```
